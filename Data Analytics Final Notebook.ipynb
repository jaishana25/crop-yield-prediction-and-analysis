Time Series Forecasting using LSTM

LSTM layers in TensorFlow require input data to follow a specific structure: (batch_size, time_steps, features). This structure is commonly used in time series data, where:

- batch_size represents the number of samples in each batch.
- time_steps represents the number of time steps in each sample.
- features represents the number of features in each time step.
Initially, input data X_train and X_test are organized in the format (samples, features). By reshaping the features to (samples, time_steps, features), we essentially create a sequence of data points where each data point contains information from previous time steps. This reshaping aligns with the requirements of LSTM models, which are designed to capture temporal dependencies in sequential data

# a)LSTM MODEL

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


X_train_ls = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_ls = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))
Y_train_ls  = Y_train.values

print(X_train_ls.shape,X_train_ls[0])
```

```
model = Sequential([
    LSTM(units=100, input_shape=(X_train_ls.shape[1], X_train_ls.shape[2])),
    Dense(units=1)
])

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_absolute_error')
model.summary()

early_stopping = EarlyStopping(patience=5, restore_best_weights=True)

history = model.fit(X_train_ls, Y_train_ls, epochs=150, batch_size=32, validation_split=0.2,
                    callbacks=[early_stopping], verbose=1)
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


Y_pred = model.predict(X_test_ls)

mse_lstm = mean_squared_error(Y_test, Y_pred)
r2_lstm = r2_score(Y_test, Y_pred)
mae_lstm = mean_absolute_error(Y_test, Y_pred)

print("Mean Absolute Error:", mae_lstm)
print("R-squared (R2):", r2_lstm)
print("Mean Squared Error:", mse_lstm)
```

```
import numpy as np

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.scatter(Y_test.values, Y_pred, color='blue', label='Actual vs Predicted')

max_val = np.max([np.max(Y_test), np.max(Y_pred)])

plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Ideal Line')

plt.title('Actual vs Predicted Yield')
plt.xlabel('Actual Yield')
plt.ylabel('Predicted Yield')
plt.legend()
plt.grid(True)
plt.show()
```

# 5.Clustering Analysis

```
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Preprocessing (handle missing values, scale numerical features, encode categorical variables)
# (Preprocessing steps depend on your dataset and its characteristics)

# Select features for clustering (replace 'selected_features' with the features you want to use)
selected_features = ['Annual_Rainfall', 'Fertilizer', 'Pesticide']

# Create feature matrix X
X = df[selected_features]

# Choose the number of clusters (replace 'n_clusters' with the desired number of clusters)
n_clusters = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Add cluster labels to the dataframe
df['Cluster'] = labels

# Visualize the clusters
plt.figure(figsize=(10, 6))
for cluster in range(n_clusters):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Annual_Rainfall'], cluster_data['Fertilizer'], label=f'Cluster {cluster}')
plt.xlabel('Annual Rainfall')
plt.ylabel('Fertilizer')
plt.title('K-means Clustering')
plt.legend()
plt.show()

# Analyze the clusters
cluster_centers = kmeans.cluster_centers_
for i, center in enumerate(cluster_centers):
    print(f"Cluster {i} Center: {center}")

# Optionally, evaluate cluster quality using metrics like silhouette score or Davies-Bouldin index
# (Evaluation depends on having ground truth labels or using internal clustering metrics)
```
